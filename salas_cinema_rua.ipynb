```python
import pandas as pd
```

```python
df_multiplex = pd.read_csv('dados/coleta_multiplex.csv')
df_multiplex
```

```python
df_salas = pd.read_csv('dados/relacao_sala_publico.csv')
df_salas['Média por Sala'] = df_salas['Média por Sala'].astype(int)
df_salas
```

```python
df_antes_2003 = df_salas[['Ano', 'Salas']].iloc[:32, :]
df_antes_2003
```

```python
df_depois_2003 = df_salas[['Ano', 'Salas']].iloc[32:, :]
df_depois_2003.loc[49, 'Salas'] = 3536
df_depois_2003 = df_depois_2003.reset_index(drop=True)
df_depois_2003
```

```python
anos = list(df_depois_2003.iloc[:, 0])
anos
```

```python
cinema_rua = df_depois_2003['Salas'].values - df_multiplex['Total_Salas'].values
df_cinema_rua = pd.DataFrame({
    'Ano': anos,
    'Qtd de salas dos cinemas de rua': cinema_rua
})

df_cinema_rua
```

```python
df_cinema_rua.to_csv('cinema_rua.csv', index=False)
```
