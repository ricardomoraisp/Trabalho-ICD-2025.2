# MEDIA PREÇO INGRESSO 2009-2024

```python
dados_2009_2015 = {
    2009: 8.61,
    2010: 9.35,
    2011: 10.13,
    2012: 11.01,
    2013: 11.73,
    2014: 12.57,
    2015: 13.59
}
dados_2016_2024 = {
    2016: 20.97,
    2017: 21.66,
    2018: 20.94,
    2019: 21.10,
    2020: 20.36,
    2021: 20.28,
    2022: 20.94,
    2023: 20.56,
    2024: 19.88
}
```

## Usando calculadora de inflação atualizamos a inflação para 2024

```python
aumento = 58.13

dados_corrigidos = {}

for ano, valor in dados_2009_2015.items():
    dados_corrigidos[ano] = round(valor + (valor * aumento / 100), 2)
```

```python
dados_finais = {}

dados_finais.update(dados_corrigidos)
dados_finais.update(dados_2016_2024)
```

## Criando csv 

```python
import pandas as pd

df = pd.DataFrame(list(dados_finais.items()), columns=["Ano", "PMI"])
df = df.sort_values("Ano")

df.to_csv("pmi_ancine.csv", index=False)
```
