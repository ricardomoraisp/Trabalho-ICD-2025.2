```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep
import os
import re
import csv
```

```python
#abre o navegador
driver = webdriver.Chrome()
driver.get('https://www.bcc.org.br/filmes')
sleep(2)
```

```python
#Define o número de páginas
paginas = 333
```

```python
#recolhe os dados
dados = []
contador_filmes = 0

for pagina in range(paginas):
    print({pagina + 1})
    #pega os links
    url_lista = f"https://www.bcc.org.br/filmes?page={pagina}"
    driver.get(url_lista)

    links = driver.find_elements(By.CSS_SELECTOR, "p.nome-filme a")
    urls = [link.get_attribute("href") for link in links]

    for url in urls:
        driver.get(url)

        # título
        sleep(2)
        titulos = driver.find_element(By.CSS_SELECTOR, "h1.title").text

        # data
        try:
            datas = driver.find_element(By.XPATH, '//p[strong[contains(text(),"Data")]]').text
            datas = datas.replace("DATA:", "").strip()
            datas = datas[:4]
        except:
            datas = ""

        dados.append([titulos, datas])

        print(f"{contador_filmes + 1} {titulos} - {datas}")
        contador_filmes += 1
```

```python
#Salva no csv
with open("filmes.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["titulo", "data"])
    writer.writerows(dados)
```
