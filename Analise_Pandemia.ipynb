# O Impacto da Pandemia no Cinema Nacional

### Desenvolvimento da Bilheteria, Nacional x Internacional

Durante o periodo da pandemia da Covid-19, o crescimento da bilheteria de filmes internacionais, embora não tenho chegado a níveis negativos, sofreu uma queda razoável, mas, com o fim do periodo supracitado, rapidamente tornou a crescer.

Nesse viés, esperava-se um comportamento similar por parte do cinema nacional, todavia o crescimento da bilheteria de filmes nacionais, embora também não tenha caído a níveis negativos, além de sofrer um impacto mais intenso durante o periodo de pandemia, mesmo após intenso investimento na Agencia do Cinema Nacional e no FSA(Fundo Setorial do Audiovisual) durante os anos subsequentes à pandemia(2023-2025), voltou a aumentar de forma inconsistente, principalmente em comparação à bilheteria internacional, sugerindo um impacto mais intenso no desenvolvimento e interesse relativos ao cinema nacional.
