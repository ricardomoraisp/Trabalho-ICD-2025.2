# O Impacto da Pandemia no Cinema Nacional

    Desenvolvimento da Bilheteria, Nacional x Internacional

        Durante o periodo da pandemia da Covid-19, o crescimento da bilheteria de filmes internacionais sofreu uma leve queda, mantendo-se razoavelmente estável, mas, com o fim do periodo supracitado, rapidamente retornou ao padrão de crescimento linear em que se encontrava. 

        Nesse viés, esperava-se um comportamento similar por parte do cinema nacional, todavia a bilheteria de filmes nacionais, além de sofrer um impacto mais intenso durante o periodo de pandemia, somente voltou a crescer( de forma menos intensa, se comparado ao periodo pré-pandemia ), após intenso investimento na Agencia do Cinema Nacional e no FSA(Fundo Setorial do Audiovisual) durante os anos subsequentes a pandemia(2023-2025).
