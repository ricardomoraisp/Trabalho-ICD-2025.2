```python
%load_ext autoreload
%autoreload 2

import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd

from constantes import (
    COLETA_MEDIA_HABITANTES_2020_2007,
    COLETA_MEDIA_HABITANTES_2006_2002,
    COLETA_MEDIA_HABITANTES_2001
)

from funcoes import coleta_media_habitantes

# Configura o navegador
options = webdriver.ChromeOptions()
meu_navegador = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

for ano in range(2020, 2000, -1):

    # Define qual texto usar baseado no ano
    if ano >= 2011:
        texto_busca = COLETA_MEDIA_HABITANTES_2020_2007
    elif ano >= 2002:
        texto_busca = COLETA_MEDIA_HABITANTES_2006_2002
    else:
        texto_busca = COLETA_MEDIA_HABITANTES_2001
    
    df_temp = coleta_media_habitantes(meu_navegador, texto_busca, ano)

    try:
        if df_temp is not None:

            df_temp = df_temp.iloc[7:36, 1:5]

            df_temp.to_csv(f'media_habitantes_por_sala_{ano}.csv', index=False)
        
    except Exception as erro_real:
        print(f"Ano {ano} falhou. O verdadeiro erro é: {erro_real}")
```
