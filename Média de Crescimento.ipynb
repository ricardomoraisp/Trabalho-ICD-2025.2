```python
import pandas
```

```python
df_bilheteria = pandas.read_csv("df_bilheteria.csv")
```

```python
df_bilheteria
```

```python
diferenca_internacional = []
diferenca_nacional = []
for i in range(11):

    diferenca_internacional.append(df_bilheteria["bilheteria_internacional"][i+1] - df_bilheteria["bilheteria_internacional"][i])
    diferenca_nacional.append(df_bilheteria["bilheteria_nacional"][i+1] - df_bilheteria["bilheteria_nacional"][i])


counter = 0
dfs_crescimento = []
for ano in range(14,25):

    d = {

            "ano" : [str(ano) + "-" + str(ano+1)],
            "Crescimento Internacional" : [diferenca_internacional[counter]],
            "Crescimento Nacional" : [diferenca_nacional[counter]]

        }
    
    dfs_crescimento.append(pandas.DataFrame(data=d))

    counter +=1
    

df_crescimento_tratado = pandas.concat(dfs_crescimento)

df_crescimento_tratado.to_csv("df_crescimento.csv")
```
