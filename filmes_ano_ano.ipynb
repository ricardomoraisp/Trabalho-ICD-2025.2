```python
import pandas as pd
```

```python
arquivo = 'filmes.csv'
arquivo2 = 'ancine1995-2024.xlsx'
df = pd.read_csv(arquivo)
df2 = pd.read_excel(arquivo2)
df.head()
```

```python
df2.head()
```

```python
df["data"] = df["data"].fillna(0).astype(int)
df["data"] = df["data"].astype(int)
df2 = df2.drop(["Certificado de Produto Brasileiro (CPB)", "Direção", "Gênero", "Empresa Produtora Brasileira Majoritária", "UF", "UF2", "Distribuidora", "Máximo de Salas", "Público acumulado", "Renda (R$) acumulada"], axis=1)

```

```python
df2.head()
```

```python
df["data"].value_counts()
```

```python
df_final = pd.concat([df, df2], ignore_index=True)
df_final
```

```python
df_final = df_final.drop_duplicates()
```

```python
df_final
```

```python
df_final["data"].value_counts()
```

```python
quantidade_df = df_final["data"].value_counts().reset_index()
quantidade_df.columns = ["data", "quantidade"]
quantidade_df
```

```python
quantidade_df.to_csv("quantidade_ano.csv", index=False)
```
