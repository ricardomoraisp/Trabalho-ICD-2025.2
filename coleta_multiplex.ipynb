### Coleta da tabela sobre Salas Multiplex

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd

from constantes import (
    NOMES_TABELAS_2020_2011,
    NOMES_TABELAS_2010_2003,
)

from funcoes import coleta_dados_multiplex

# Configura o navegador
options = webdriver.ChromeOptions()
meu_navegador = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# Cria uma lista vazia para ir guardando as tabelas
lista_de_tabelas = []

for ano in range(2020, 2002, -1):
    
    # Define qual texto usar baseado no ano
    if ano >= 2011:
        texto_busca = NOMES_TABELAS_2020_2011
    elif ano >= 2007:
        texto_busca = NOMES_TABELAS_2010_2003

    # Coleta a tabela do ano específico
    df_temp = coleta_dados_multiplex(meu_navegador, texto_busca, ano)
    
    # Se a tabela foi coletada com sucesso (não é None)
    if df_temp is not None:
        
        # Fazemos a limpeza e adicionamos o ano *antes* de guardar
        # Fica só com as duas primeiras colunas
        if ano >= 2011:
            df_temp = df_temp.iloc[:, :3].copy()
            df_temp = df_temp.drop(df_temp.columns[1], axis=1)
        else:
            df_temp = df_temp.iloc[:, :2].copy()
        
        # Adiciona o ano. Usar colunas[0] é mais seguro que iloc para sobrescrever.
        nome_primeira_coluna = df_temp.columns[0]
        df_temp[nome_primeira_coluna] = ano 
        
        # Guarda a tabela pronta na lista
        lista_de_tabelas.append(df_temp)
        print(f"Tabela de {ano} guardada com sucesso!")
        
    else:
        print(f"ATENÇÃO: Falha ao coletar dados para o ano {ano}.")

if len(lista_de_tabelas) > 0:
    # O pd.concat engole a lista inteira de uma vez e cria o DataFrame final
    df_final = pd.concat(lista_de_tabelas, ignore_index=True)
    
    # Ajusta os cabeçalhos finais
    df_final.columns = ['Ano', 'Total_Salas', 'Total_Salas_2']
    
    print("\n--- Tabela final consolidada: ---")
else:
    print("\nFalha crítica: Nenhuma tabela foi coletada para montar o DataFrame final.")

meu_navegador.quit()
```

Preenche os NaNs de Total_Salas com os valores da coluna 'Total_Salas_2 '

```python
df_final['Total_Salas'] = df_final['Total_Salas'].fillna(df_final['Total_Salas_2'])

df_final = df_final.drop(columns=['Total_Salas_2'])
```

Ordena os anos em ordem crescente e os define como index da função.

```python
df_final = df_final.sort_values(by=['Ano', 'Total_Salas'], ascending=True)
df_final.set_index('Ano', inplace=True)
```

Define o tipo de dado da coluna 'Total_Salas' como inteiro.

```python
df_final['Total_Salas'] = df_final['Total_Salas'].str.replace('.', '').astype(int)
df_final
```
